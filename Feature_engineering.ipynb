```python
player_data = pd.read_csv("player_data.csv" )
player_data.info()
```

```python
#player_data.fillna(0,  inplace = True)
player_data
```

```python
player_data.fillna( method ="pad", inplace = True)
```

```python
player_data.info()
```

```python
#drop 'birth_date'
player_data['birth_years'] = player_data['birth_date'].apply(lambda x: x[-4:])             
player_data['birth_day'] = player_data['birth_date'].apply(lambda x: x.split(' ')[1].strip(','))  
player_data['birth_months'] = player_data['birth_date'].apply(lambda x: x.split(' ')[0])  
player_data.drop(['birth_date'], axis=1, inplace = True)
```

```python
player_data
```

```python
#https://docs.python.org/2/library/datetime.html#strftime-strptime-behavior

from time import strptime
import datetime

#month_number = datetime.datetime.strptime(month_name, '%b').month
player_data['birth_months'] = player_data['birth_months'].apply(lambda x: datetime.datetime.strptime(x, '%B').month)  
```

```python
#how many years been incollege
player_data['college_years'] = player_data['year_end'] - player_data['year_start']
```
