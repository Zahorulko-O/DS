**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
# Ваш код здесь
#1
#print(data['marital-status'].value_counts())
marital_status = data.loc[:, ('marital-status', 'salary')]

married_unmarried = {'Married-civ-spouse': 'married', "Married-spouse-absent": 'married', 'Married-AF-spouse': 'married', 
                "Never-married": "unmarried", 'Divorced': 'unmarried', 'Separated': 'unmarried', 'Widowed': 'unmarried'}
marital_status['marital-status'] = marital_status['marital-status'].map(married_unmarried) 
sal = {'<=50K': 0, '>50K': 1}
marital_status['salary'] = marital_status['salary'].map(sal)

a = marital_status.pivot_table(['salary'], ['marital-status'], aggfunc=(np.sum))
#print(a)
if a['salary'][0] > a['salary'][1]:
    print("Зарабатывающих много больше среди женатых")
else:
    print("Among whom there is more than a lot of earners is single")
```

```python
#2

data['N'] = 1
marital = {'Married-civ-spouse': 1, 'Married-spouse-absent': 1, 'Married-AF-spouse': 1, 
           'Divorced': 0, 'Never-married': 0, 'Widowed': 0, 'Separated': 0}  # 1 - женатыe; 0 - холостые


data
d1 = data.drop(['age', 'workclass', 'fnlwgt', 'education', 'education-num', 
                'occupation', 'relationship', 'race', 'sex', 
                'capital-gain', 'capital-loss', 'hours-per-week', 'native-country'], axis=1)
d1['marital-status'] = d1['marital-status'].map(marital)

#a = math.fabs(d1[(d1['marital-status'] == 1) & (d1['salary'] == 'True')]['N'].sum()) #женатыe
a = d1[(d1['marital-status'] == 1) & (d1['salary'] == '>50K')]['N'].sum()
b = d1[(d1['marital-status'] == 0) & (d1['salary'] == '>50K')]['N'].sum() #холостые

#print(' Среди кого больше доля зарабатывающих много: ', np.where(a>b, "женатыx", "холостыx"))
print("Among whom there is more than a lot of earners: ", np.where(a>b,"married", "single"))
```

**10. Посчитайте среднее время работы (*hours-per-week*) зарабатывающих мало и много (*salary*) для каждой страны (*native-country*).**

```python
# Ваш код здесь
#1
mean_per = data.groupby(['salary', 'native-country'])['hours-per-week'].agg([np.mean])
print('Cреднее время работы людей, зарабатывающих мало: ', '\n', 
      round(mean_per.loc[[('<=50K'),  'native-country']], 2))
print('\n\n', 'Cреднее время работы людей, зарабатывающих много: ', '\n', 
      round(mean_per.loc[[('>50K'),  'native-country']], 2))
```

```python
#2
salary_= data.pivot_table(['hours-per-week'], ['salary', 'native-country'], aggfunc='mean')
print('Cреднее время работы людей, зарабатывающих мало: ')
print(salary_.loc[[('<=50K')]])
print('\n\n', 'Cреднее время работы людей, зарабатывающих много:')
print(salary_.loc[[('>50K')]])
```
