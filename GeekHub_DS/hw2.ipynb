```python
import numpy as np

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris = np.genfromtxt(url, delimiter=',', dtype='object')
names = ('sepallength', 'sepalwidth', 'petallength', 'petalwidth', 'species')
np.__version__
```

```python
print(iris)
```

```python
##1 Извлечь колонку ‘species’. 2 Преобразовать первые 4 колонки в 2D массив

b = np.array(iris[:,0:4])
print(b)
```

```python
## 3 Посчитать mean, median, standard deviation по 1-й колонке
b = np.array(b, dtype='float')

print("Mean value is: ", b[:, 0:1].mean())
print("Mean value is: ", np.mean(b[:, 0:1]))
```

```python
print("Median value is: ", np.median(b[:, 0:1]))
```

```python
print("Mean standard deviation is: ", b[:, 0:1].std())
```

```python
## 4 Вставить 20 значений np.nan на случайные позиции в массиве

np.put(b, np.random.choice(range(150*4), 20, replace=False), np.nan)
print(b)
b.dtype.name
```

```python
 ### 5 Найти позиции вставленных значений np.nan в 1-й колонке
c = np.array((b[:,0:1]), dtype='float64')
np.where(np.isnan(c))
```

```python
## 6 Отфильтровать массив по условию значения в 3-й колонке > 1.5 и значения в 1-й колонке < 5.0
more1_5 = np.where(b[:,3] > 1.5)
print("Positions where value > 1.5: ", more1_5)
```

```python
less_5 = np.where(b[:,1] < 5)
print("Positions where value < 5: ", less_5)
```

```python
## 7 Заменить все значения np.nan на 0
missing_bool = np.isnan(b) | np.isinf(b)
b[missing_bool] = 0 
print(b)
```

```python
### 8 Посчитать все уникальные значения в массиве и вывести их вместе с подсчитанным количеством
uniqs, counts = np.unique(b, return_counts=True)
print("Unique items : ", uniqs)
print("\n")
print("Counts       : ", counts)
```

```python
## 9. Разбить массив по горизонтали на 2 массива
b2 = np.array_split(b,2)
print(b2)
np.array
```

```python
b_1 = np.array(b[0:76,:])
print(b_1)
b_2 = np.array(b[76:,:])
print(b_2)
```

```python
##10 Отсортировать оба получившихся массива по 1-й колонке: 1-й по возрастанию, 2-й по убывани.
sorted_index_1stcol = b_1[:, 0].argsort()
print(b_1[sorted_index_1stcol])
```

```python
sorted_index_2stcol = b_2[:, 0].argsort()
print(b_2[sorted_index_2stcol[::-1]])
```

```python
##11 Склеить оба массива обратно
b = np.vstack((b_1,b_2))
print(b)
```

```python
##12 Найти наиболее часто повторяющееся значение в массиве

uniqs, counts = np.unique(b, return_counts=True)
print(uniqs[np.argmax(counts)])
```
