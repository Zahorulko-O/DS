Избавимся сразу от переменных, названия которых заканчиваются на `_lognorm` (нужны для соревнования на Kaggle). Выберем их с помощью `filter()` и удалим `drop`-ом:

```python
df.drop(filter(lambda c: c.endswith('_lognorm'), df.columns), 
        axis = 1,       # axis = 1: столбцы 
        inplace = True) # избавляет от необходимости сохранять датасет
```

Столбец **`published`** (время публикации) содержит строки. Чтобы мы могли работать с этими данными как с датой/временем публикации, приведём их к типу `datetime`:

```python
print(df.published.dtype)
df['published'] = pd.to_datetime(df.published, yearfirst = True)
print(df.published.dtype)
```

Создадим несколько столбцов на основе данных о времени публикации:

```python
df['year'] = [d.year for d in df.published]
df['month'] = [d.month for d in df.published]

df['dayofweek'] = [d.isoweekday() for d in df.published]
df['hour'] = [d.hour for d in df.published]
```

## 3\. Когда лучше всего публиковать статью?

* Больше всего просмотров набирают статьи, опубликованные в 12 часов дня
* У опубликованных в 10 утра постов больше всего комментариев
* Больше всего просмотров набирают статьи, опубликованные в 6 часов утра
* Максимальное число комментариев на гиктаймсе набрала статья, опубликованная в 9 часов вечера
* На хабре дневные статьи комментируют чаще, чем вечерние

```python
#### *** #### Больше всего просмотров набирают статьи, опубликованные в 12 часов дня
#(df.groupby('hour')['views'].mean().sort_values(ascending=False)[:10])
#### *** #### 

#df.groupby('hour')['views'].mean().sort_values(ascending=False)[:10].plot(kind='bar', color='hotpink', rot=90,  fontsize=10);
df_=df.groupby('hour')[['views']].mean().sort_values(by='views', ascending=False)
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)    
    
ax.vlines(x=df_.index, ymin=0, ymax=df_['views'], color='pink', alpha=0.7, linewidth=2)
ax.scatter(x=df_.index, y=df_.views, s=75, color='hotpink', alpha=0.7)


plt.ylabel('Количество', fontdict = {'size':20})# , loc = 'center' )
plt.xlabel('Часы публикаций' , fontdict = {'size':20} )#, loc = 'left')

print('Cтатьи, опубликованные в 12 часов дня не набирают больше всего просмотров набирают')
print("Больше всего просмотров набирают статьи, опубликованные в 6 часов утра")
```

```python
#У опубликованных в 10 утра постов больше всего комментариев
df_g = df.groupby('hour')[['comments']].mean().sort_values(by ='comments', ascending=False)
# Draw plot
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)    
    
ax.vlines(x=df_g.index, ymin=0, ymax=df_g.comments, color='firebrick', alpha=0.7, linewidth=2)
ax.scatter(x=df_g.index, y=df_g.comments, s=75, color='firebrick', alpha=0.7)
# Title, Label, Ticks and Ylim
ax.set_title('Когда лучше всего публиковать статью', fontdict={'size':22})
ax.set_ylabel('Коментарии')
ax.set_xticks(df_g.index)
plt.show()
print("У опубликованных не в 10 утра постов больше всего комментариев")
```

```python
#Максимальное число комментариев на гиктаймсе набрала статья, опубликованная в 9 часов вечера
#На хабре дневные статьи комментируют чаще, чем вечерние
#df.loc[df['domain'] == 'geektimes.ru'].groupby('hour')[['comments']].mean().plot();
#df.loc[df['domain'] == 'habrahabr.ru'].groupby('hour')[['comments']].mean().plot();
#
plt.figure(figsize=(20,10))
plt.plot(df.loc[df['domain'] == 'geektimes.ru'].groupby('hour')[['comments']].mean(), color="dodgerblue", label="geektimes.ru");
plt.plot(df.loc[df['domain'] == 'habrahabr.ru'].groupby('hour')[['comments']].mean(), color="orange", label="habrahabr.ru");

plt.ylabel('Количество', fontdict = {'size':16})
plt.xlabel('Часы публикаций' , fontdict = {'size':16})
print("Максимальное число комментариев на гиктаймсе набрала статья, опубликованная не в 9 часов вечера")
print("На хабре дневные статьи не комментируют чаще, чем вечерние")
```
